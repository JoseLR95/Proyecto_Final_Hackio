```python
from statsbombpy import sb
import pandas as pd
import numpy as np
import time
```

```python
competiciones = sb.competitions()
df_LaLiga = competiciones[competiciones["competition_id"] == 11]
df_partidos = pd.DataFrame()
for x in df_LaLiga["season_id"]:
    df_temporal = sb.matches(competition_id=11, season_id = x)
    df_partidos = pd.concat([df_partidos, df_temporal], ignore_index=True)
liga2015 = df_partidos[df_partidos["season"] == "2015/2016"]
lista_2015 = []
for x in liga2015["match_id"]:
    lista_2015.append(x)
df_eventos_liga2015 = pd.DataFrame()
errores = []
for x in lista_2015:
    try:
        evento= sb.events(match_id=x)
        df_temp = evento
        df_eventos_liga2015 = pd.concat([df_eventos_liga2015, df_temp], ignore_index=True)
        time.sleep(0.5)
        time.sleep(0.5)
    except Exception as e:
        errores.append(x)
df_eventos_liga2015.to_csv("../Data/df_eventos_liga2015.csv", index=False)
```

```python
lista_2015_todos = df_eventos_liga2015['match_id'].unique().tolist()
```

```python
lista = liga2015['match_id'].unique().tolist()
```

```python
faltan = []
for x in lista:
    if x not in lista_2015_todos:
        faltan.append(x)
faltan
```

```python
df_eventos_liga2015_faltantes = pd.DataFrame()
for x in faltan:
    evento= sb.events(match_id=x)
    df_temp = evento
    df_eventos_liga2015_faltantes = pd.concat([df_eventos_liga2015_faltantes, df_temp], ignore_index=True)
```

```python
df_eventos_liga2015 = pd.concat([df_eventos_liga2015, df_eventos_liga2015_faltantes], ignore_index=True)
```

```python
df_eventos_liga2015.to_csv("data/df_eventos_liga2015.csv", index=False)
```
